```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# 이미지 불러오기
img = cv2.imread('/content/test2.jpeg')

# 다각형 좌표 정의
polygon1 = np.array([
    [80, 330],  # 왼쪽 상단 꼭짓점
    [180, 330],  # 오른쪽 상단 꼭짓점
    [180, 485],  # 오른쪽 하단 꼭짓점
    [80, 485]   # 왼쪽 하단 꼭짓점
], np.int32)

polygon2 = np.array([
    [230, 450],  # 왼쪽 상단 꼭짓점
    [500, 450],  # 오른쪽 상단 꼭짓점
    [500, 700],  # 오른쪽 하단 꼭짓점
    [230, 700]   # 왼쪽 하단 꼭짓점
], np.int32)

# 다각형을 이미지 위에 그리기
cv2.polylines(img, [polygon1], isClosed=True, color=(0, 255, 0), thickness=2)
cv2.polylines(img, [polygon2], isClosed=True, color=(0, 255, 0), thickness=2)

# 결과 이미지 표시
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
