```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
img = cv2.imread('/content/test2.jpeg')

# Îã§Í∞ÅÌòï Ï¢åÌëú Ï†ïÏùò
polygon1 = np.array([
    [80, 330],  # ÏôºÏ™Ω ÏÉÅÎã® Íº≠ÏßìÏ†ê
    [180, 330],  # Ïò§Î•∏Ï™Ω ÏÉÅÎã® Íº≠ÏßìÏ†ê
    [180, 485],  # Ïò§Î•∏Ï™Ω ÌïòÎã® Íº≠ÏßìÏ†ê
    [80, 485]   # ÏôºÏ™Ω ÌïòÎã® Íº≠ÏßìÏ†ê
], np.int32)

polygon2 = np.array([
    [230, 450],  # ÏôºÏ™Ω ÏÉÅÎã® Íº≠ÏßìÏ†ê
    [500, 450],  # Ïò§Î•∏Ï™Ω ÏÉÅÎã® Íº≠ÏßìÏ†ê
    [500, 700],  # Ïò§Î•∏Ï™Ω ÌïòÎã® Íº≠ÏßìÏ†ê
    [230, 700]   # ÏôºÏ™Ω ÌïòÎã® Íº≠ÏßìÏ†ê
], np.int32)

# Îã§Í∞ÅÌòïÏùÑ Ïù¥ÎØ∏ÏßÄ ÏúÑÏóê Í∑∏Î¶¨Í∏∞
cv2.polylines(img, [polygon1], isClosed=True, color=(0, 255, 0), thickness=2)
cv2.polylines(img, [polygon2], isClosed=True, color=(0, 255, 0), thickness=2)

# Í≤∞Í≥º Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
